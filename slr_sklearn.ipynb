```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("placement.csv")
df.shape
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.xlabel("cgpa")
plt.ylabel("package")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["package"]), df.drop(columns=["cgpa"]), test_size = 0.2, random_state=2)
```

```python
# import simple linear regression closed form solution model

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.predict(X_test.values[0].reshape(-1,1))
```
